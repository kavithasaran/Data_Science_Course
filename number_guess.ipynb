## 1. print if user guessed number is right or not

```python
secret_num = 56
for i in range(0,7):
    user_input = int(input("Enter your guess"))
    
    if secret_num == user_input:
        print("congratulations! your guess is right")
        break
    else:
        print("sorry your guess is wrong")
```

## while loop

```python
import random

print("welcome to the game")
print("guess any number between 1 to 100")

secret_number = random.randint(1,100)
i = 1
while True:
    user_input = int(input("Enter your guess"))
    if user_input < 1 or user_input > 100:
        print("Invalid Number")
        i = i + 1
    elif user_input < secret_number:
        print("your number is lesser")
        i = i + 1
    elif user_input > secret_number:
        print("your number is higher")
        i = i + 1
    else:
        print("your guessed number is correct")
        break
if user_input == secret_number:
    print(f"congrats!, you have guessed the number {secret_number} in {i} attempts")
else:
    print("all your tries are over")
```

## incldue exceptional condition

```python
import random

print("welcome to the game")
print("guess any number between 1 to 100")

secret_number = random.randint(1,100)
i = 1
for i in range(1,8):
    try:
        user_input = int(input("Enter your guess"))
        if user_input < 1 or user_input > 100:
            print("Invalid Number")
            i = i + 1
        elif user_input < secret_number:
            print("your number is lesser")
            i = i + 1
        elif user_input > secret_number:
            print("your number is higher")
            i = i + 1
        else:
            print("your guessed number is correct")
            break
    except ValueError:
        print("Please enter only number")
if user_input == secret_number:
    print(f"congrats!, you have guessed the number {secret_number} in {i} attempts")
else:
    print("all your tries are over")
       
```
