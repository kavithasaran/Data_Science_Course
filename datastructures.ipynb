# data structures - python
* to store or get multiple data or values in built in structures

## 4 types of data structures:
1. list
2. set
3. tuple
4. libraries

### list

* accepts duplicate values
* single dimension
* inside square bracket
* heterogenous data accepted - string, int, float
* indexing for the position automatically done  - strat with 0 from lft - start with -1 from right
* slicing can be done 0:2 - fetches values in 0,1 poistion always excluding the last positin number(here 2) - so values in 0 and 1 are displayed
* mutable - values in any position can be changed in future (marks[2] = 23) this will update the values 

```python
stu_marks = [54,76,87,98,90]
stu_marks
```

```python
type(stu_marks) ##returns the type of data structures it is stored
```

```python
stu_marks[-4:-1] ##fetch particular index value & slicing
```

```python
stu_marks[3] = 100
```

```python
stu_marks
```

### set

* accepts only unique values, duplicates truncated automatically
* single dimension
* inside curly braces
* heterogenous data accepted - string, int, float
* indexing and slicing is not possible
* mutable - values can be added or removed thru functions()

```python
empty_set = set() # create an empty set
```

```python
type(empty_set)
```

```python
items_to_buy = {"apple","grape","guava","pomo","melon",45,2.3,"guava"}
```

```python
type(items_to_buy)"apple","grape","guava","pomo","melon",45,2.3,"guava"
```

```python
items_to_buy.add("banana")
```

```python
items_to_buy
```

## dictionary - key and value
* its a key:value pair
* two dimensional values
* inside curly braces with key:value 
* heterogenous data accepted - string, int, float
* keys cannot be duplicate, values can be repeated for diff keys
* if duplicate keys given, latest value given to the keys are updated and previous values are erased
* indexing and slicing cannot be done
* but the keys and values are fetched seperately thru functions()

```python
empty_dict = {}
```

```python
type(empty_dict)
```

```python
empty_dict = {'ram':[34,34,21,54],'ravi':[32,54,65,76],'priya':[43,65,76,87],'ravi':[33,43,76,87]}
```

```python
type(empty_dict)
```

```python
empty_dict
```

```python
empty_dict[0:2] ## slicing cant be done
```

```python
empty_dict.update([ravi, ] 33)
```

```python
import pandas as pd
pd.DataFrame(empty_dict,index=None,columns=None,dtype=None,copy=None)
```
