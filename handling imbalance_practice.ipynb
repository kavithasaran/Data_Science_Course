# 1. Import neccesary libraries

```python
from sklearn import datasets
```

# 2. Import Dataset

```python
breast_cancer_data = datasets.load_breast_cancer()
```

# 3. data understanding

```python
print(breast_cancer_data.DESCR)
```

```python
breast_cancer_data.feature_names
```

```python
breast_cancer_data.target
```

# 4.Data preparation

```python
import pandas as pd
```

```python
BC_dataframe = pd.DataFrame(data = breast_cancer_data.data, columns = breast_cancer_data.feature_names )
BC_dataframe["Target"] = breast_cancer_data.target
BC_dataframe
```

```python
X = BC_dataframe.drop(labels = "Target", axis=1)
X
```

```python
y = BC_dataframe["Target"]
y
```

```python
import matplotlib.pyplot as plt
```

```python
y.value_counts()
```

```python
plt.pie(x = y.value_counts(),labels = ["Benign cancer","Malignant cancer"],autopct = "%1.0f%%")
plt.show()
```

## NOW to handle the imbalance data, scalar techniques are used to change the input feature's values into mean and std.deviation

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scalar = MinMaxScaler()
scaled_X = pd.DataFrame(min_max_scalar.fit_transform(X),columns = breast_cancer_data.feature_names )
scaled_X
```

```python
X.mean(), X.std()
```

```python
scaled_X.mean(), scaled_X.std()
```

# 5. model building,

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y,test_size = 0.20,random_state = 123,shuffle=True,
    stratify=y)
```

```python
y_train.value_counts()
```

```python
plt.pie(x=y_train.value_counts(),autopct="%1.0f%%")
plt.show()
```

```python
plt.pie(x=y_test.value_counts(),autopct="%1.0f%%")
plt.show()
```

# 6.MODEL training

```python
%%time 
from sklearn.tree import DecisionTreeClassifier
DT_Model = DecisionTreeClassifier(criterion='gini',max_depth=3,class_weight={0:2,1:1})
DT_Model
```

```python
DT_Model.fit(X_train,y_train)
```

# model testinf and evaluation

```python
y_pred_train = DT_Model.predict(X_train)
y_pred_train
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_pred_train,y_train))
```

```python
print(classification_report(y_pred_train,y_train))
```

```python
y_pred_test = DT_Model.predict(X_test)
y_pred_test
```

```python
print(confusion_matrix(y_pred_test,y_test))
```

```python
print(classification_report(y_pred_test,y_test))
```
